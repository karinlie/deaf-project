```python
from ultralytics import YOLO
import cv2

# Last inn modellen
model = YOLO('yolov8n.pt')

# Kj√∏r YOLO p√• webkamera (source=0)
results = model(source=0, conf=0.3)

# G√• gjennom resultatene og vis bildene manuelt
for result in results:
    frame = result.plot()  # F√• bildet med tegnet bounding box
    cv2.imshow('YOLO Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Trykk 'q' for √• avslutte
        break

cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

for result in results:
    frame = result.plot()
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
